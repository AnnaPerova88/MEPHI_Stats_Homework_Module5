Вывод о целесообразности использования корреляции Пирсона
На основе диаграммы рассеяния можно визуально оценить, что данные имеют линейную зависимость. Точки образуют линейный тренд, поэтому можно сделать вывод, что можно использовать **корреляцию Пирсона**. Если бы тренд был нелинейный или распределение сильно выбросное/кластерное, рекомендуется использовать другие методы, например, *коэффициент корреляции Спирмена.*

```python
# КРИТЕРИЙ 2

# расчет коэффициента Пирсона

# расчет t-критерия

# вывод о степени корреляции величин и статистической значимости их взаимосвязи
```

```python
from scipy.stats import pearsonr

# Расчёт коэффициента корреляции Пирсона
corr, p_value = pearsonr(df['x'], df['y'])

# Вывод результата
print(f'Коэффициент корреляции Пирсона: {corr:.3f}')
print(f'P-value: {p_value:.3e}')
```

*  Коэффициент корреляции Пирсона: 0.803
*  P-value: 3.974e-114

```python
import numpy as np

# Количество наблюдений
n = len(df)

# Расчёт t-статистики
t_stat = corr * np.sqrt((n - 2) / (1 - corr**2))

# Расчёт критического значения
from scipy.stats import t
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, df=n-2)

# Вывод результатов
print(f't-статистика: {t_stat:.3f}')
print(f'Критическое значение t (при уровне значимости {alpha}): {t_critical:.3f}')

# Проверка гипотезы
if abs(t_stat) > t_critical:
    print('Корреляция статистически значима.')
else:
    print('Корреляция статистически незначима.')
```

*  t-статистика: 30.092
*  Критическое значение t (при уровне значимости 0.05): 1.965
*  Корреляция статистически значима.

```python
# КРИТЕРИЙ 4

# расчет коэффициента Спирмена

# расчет t-критерия

# вывод о степени корреляции величин и статистической значимости их взаимосвязи
```

```python
from scipy.stats import spearmanr

# Расчёт коэффициента Спирмена
corr, p_value = spearmanr(df2['x'], df2['y'])

# Вывод результата
print(f'Коэффициент корреляции Спирмена: {corr:.3f}')
print(f'P-value: {p_value:.3e}')
```

*  Коэффициент корреляции Спирмена: 0.617
*  P-value: 8.472e-54

```python
import numpy as np

# Количество наблюдений
n = len(df2)

# Расчёт t-статистики
t_stat = corr * np.sqrt((n - 2) / (1 - corr**2))

# Расчёт критического значения
from scipy.stats import t
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, df=n-2)

# Вывод результатов
print(f't-статистика: {t_stat:.3f}')
print(f'Критическое значение t (при уровне значимости {alpha}): {t_critical:.3f}')

# Проверка гипотезы
if abs(t_stat) > t_critical:
    print('Корреляция статистически значима.')
else:
    print('Корреляция статистически незначима.')
```

*  t-статистика: 17.500
*  Критическое значение t (при уровне значимости 0.05): 1.965
*  Корреляция статистически значима.

Значение t-статистики (17.500) значительно превышает критическое значение (1.965) при уровне значимости 0.05. Это свидетельствует о том, что взаимосвязь между переменными статистически значима, а не случайна.

# Построение диаграммы рассеяния

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение данных
df3 = pd.read_csv('sample_data/ex2.csv')

# Построение диаграммы рассеяния
plt.figure(figsize=(8, 6))
plt.scatter(df3['x'], df3['y'], alpha=0.7, label='Данные')
plt.title('Диаграмма рассеяния: x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Расчёт коэффициента корреляции Пирсона

```python
from scipy.stats import pearsonr

# Расчёт коэффициента корреляции Пирсона
corr, p_value = pearsonr(df3['x'], df3['y'])

# Вывод результатов
print(f'Коэффициент корреляции Пирсона: {corr:.3f}')
print(f'P-value: {p_value:.3e}')
```

```python
# КРИТЕРИЙ 6

# расчет коэффициентов линейной регрессии

# построение диаграммы рассеяния прогнозных и реальных значений на одном графике
```

# Расчет коэффициентов линейной регрессии

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Модель линейной регрессии
X = df3[['x']].values  # Независимая переменная (предиктор)
y = df3['y'].values    # Зависимая переменная

# Обучение модели
model = LinearRegression()
model.fit(X, y)

# Получение коэффициентов
slope = model.coef_[0]  # Коэффициент наклона
intercept = model.intercept_  # Свободный член

# Вывод результатов
print(f'Коэффициент наклона (slope): {slope:.3f}')
print(f'Свободный член (intercept): {intercept:.3f}')
```

*  Коэффициент наклона (slope): 18.557
*  Свободный член (intercept): 25.743

# Построение диаграммы прогнозных реальных значений

```python
# Прогнозные значения
y_pred = model.predict(X)

# Построение графика
plt.figure(figsize=(8, 6))
plt.scatter(df3['x'], df3['y'], label='Реальные значения', alpha=0.7)
plt.plot(df3['x'], y_pred, color='red', label='Прогнозные значения', linewidth=2)
plt.title('Реальные vs Прогнозные значения')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# КРИТЕРИЙ 7

# расчет средней ошибки аппроксимации
```

```python
# Расчёт средней ошибки аппроксимации
error = np.mean(np.abs((y - y_pred) / y)) * 100

# Вывод результата
print(f'Средняя ошибка аппроксимации: {error:.2f}%')
```

* Средняя ошибка аппроксимации: 23.12%

```python
# КРИТЕРИЙ 8

# расчет F-критерия

# вывод о статистической значимости уравнения в целом
```

```python
from sklearn.metrics import r2_score

# Коэффициент детерминации (R^2)
r2 = r2_score(y, y_pred)

# Число наблюдений и параметров
n = len(y)
k = 1  # Количество предикторов

# Расчёт F-критерия
F_stat = (r2 / k) / ((1 - r2) / (n - k - 1))

# Вывод результата
print(f'F-критерий: {F_stat:.3f}')
```

* F-критерий: 579.421
* Модель является статистически значимой

```python
import statsmodels.api as sm

# Добавление константы для модели
X_with_const = sm.add_constant(X)

# Построение модели с использованием statsmodels
model_sm = sm.OLS(y, X_with_const).fit()

# Стандартные ошибки коэффициентов
print(model_sm.summary())
```

```python
# Извлечение t-статистики
t_values = model_sm.tvalues

# Вывод t-статистики
print(f't-статистика коэффициентов: {t_values}')
```

* t-статистика коэффициентов: [56.58903227 24.07117378]
